Q2. Given a Pandas DataFrame df with columns 'A', 'B', and 'C', write a Python function to re-index the
DataFrame with a new index that starts from 1 and increments by 2 for each row.

```python
def reindex(df1):
    new_index = pd.RangeIndex(1,len(df1)*2,  2)
    df1.index  = new_index
    return df1
```

```python
data ={
    "A" : [576, 9086, 56,88],
    "B" : ["Shivangi","Nitin","Nishi","Anmol"],
    "C" : [22.33 ,12.56 ,0.08,8.4]
  
}
df1 =pd.DataFrame(data)
df1 = reindex(df1)
df1
```

Q5. How are DataFrame.size() and DataFrame.shape() different?

DataFrame.size returns the total number of elements in the DataFrame as a single integer.

DataFrame.shape returns a tuple with the number of rows and columns as separate values.

DataFrame.size does not use parentheses when accessed, while DataFrame.shape is accessed with parentheses.

DataFrame.size provides more granular information about the size of the DataFrame, whereas DataFrame.shape provides information about its dimensions.

Q8. You have a Pandas DataFrame df with columns 'A', 'B', and 'C'. Write a Python function that selects
all rows where the value in column 'A' is greater than 5 and the value in column 'B' is less than 10. The
function should return a new DataFrame that contains only the selected rows.

```python
def select_rows(df8):
    selected_row = df[(df["A"] > 5) & (df["B"] < 10)]
    return selected_row
```

Q10. Given a Pandas DataFrame df with a column 'Sales' and a column 'Date', write a Python function to

for each row in the DataFrame. The moving average should be calculated using a window of size 7 and
should include the current day.

```python
data10 = {
    "Date" : pd.date_range("01/01/2023",periods =30),
    "Sales" : [10, 20, 15, 25, 30, 35, 40, 50, 45, 55, 60, 70, 80, 75, 65, 55, 45, 50, 40, 30, 25, 20, 15, 10, 5, 15, 25, 35, 45, 55]
}
df10 = pd.DataFrame(data10)
df10["MovingAverage"] = df["Sales"].rolling(window =7, min_periods = 1).mean()
df10
```

Q12. Given a Pandas DataFrame df with a column 'Date' that contains timestamps, write a Python
function to select all rows where the date is between '2023-01-01' and '2023-01-31'.

```python
def selected_dates(df12):
    start_date = '2023-01-01'
    end_date = '2023-01-31'
    selected_rows = df[(df12["Date"]>=start_date) & (df12["Date"]<=end_date)]
    print( selected_rows)


data12 = {
    "Date": ['2023-01-05', '2023-01-15', '2023-02-10', '2023-01-25', '2023-03-05']
}
df12= pd.DataFrame(data12)
df12  = selected_dates(df12)
df12
```
